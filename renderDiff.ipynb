```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
origin = pd.read_csv('runs/detect/train/results.csv')
origin.head()
```

```python
transformer = pd.read_csv('runs/detect/train2/results.csv')
transformer.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Precision performance between origin yolov8 and transformered yolov8')
plt.plot(origin['   metrics/precision(B)'],label='yolov8_origin')
plt.plot(transformer['   metrics/precision(B)'],label='yolov8_transformer')
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.legend()

plt.savefig('precision.png',dpi=300)
plt.show()
```

```python
import os
# get the file size
def get_file_size(file_path):
    size = os.path.getsize(file_path)
    return size

# get the file size
origin_size = get_file_size('./yolov8n.pt')
transformer_size = get_file_size('./yolov8n_transformer.pt')
```

```python
plt.figure(figsize=(10,5))
plt.title('Model size comparison between origin yolov8 and transformered yolov8')
plt.bar("yolov8",origin_size, label='yolov8_origin')
plt.bar("yolov8_transformer",transformer_size, label='yolov8_transformer')
plt.legend()
plt.xlabel('Model')
plt.ylabel('Size(MiB)')
plt.text(0,origin_size+0.1,str(round(origin_size / 1024 / 1024,2))+"MiB",ha='center',va='bottom',fontsize=10)
plt.text(1,transformer_size-10,str(round(transformer_size / 1024 / 1024,2))+"MiB",ha='center',va='bottom',fontsize=10)
plt.savefig('size.png',dpi=300)
plt.show()
```
